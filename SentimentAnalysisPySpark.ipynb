```python
from textblob import TextBlob
from pyspark.sql.functions import from_unixtime, to_date, year, udf

sentiment = udf(lambda x: TextBlob(x).sentiment[0])
spark.udf.register("sentiment_tb", sentiment)
df_final = df_final.withColumn('sentiment_tb',sentiment('reviews').cast('double'))
df_final.show()
```

```python
# vader model

from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer

def getSentimentScore(Text):
    analyzer = SentimentIntensityAnalyzer()
    vs = analyzer.polarity_scores(Text)
    return float(vs['compound'])
```

```python
udfSentimentScore = udf(getSentimentScore)
df_final = df_final.withColumn('sentiment_vd', udfSentimentScore('reviews'))
df_final.select('Product_ID','reviews','sentiment_tb')
df_final.show()
```
