```python
# translate all the reviews to english
# language detection and translation
# loop through the reviews column and pass each list of reviews to the cleaning function

from langdetect import detect
from deep_translator import GoogleTranslator

import re
import nltk
nltk.download('stopwords')
stemmer = nltk.SnowballStemmer("english")  # to convert words to root words
from nltk.corpus import stopwords   
import string
stopword=set(stopwords.words('english')) #english stop word dictionary

def clean(text):
    lang = detect(text)
    if lang == 'en':
        text = str(text).lower()   #convert to lower case 
        text = re.sub('\[.*?\]', '', text) # remove punctuations 
        text = re.sub('@[^\s]+','',text)  #remove tagged user
        text = re.sub('https?://\S+|www\.\S+', '', text) #remove URLs
        text = re.sub('<.*?>+', '', text) 
        text = re.sub('[%s]' % re.escape(string.punctuation), '', text)
        text = re.sub('\n', '', text)
        text = re.sub('\w*\d\w*', '', text)
        text = [word for word in text.split(' ') if word not in stopword] # word toeknization and removal of stop words
        text=" ".join(text) #join words again to form string
        text = [stemmer.stem(word) for word in text.split(' ')] # stemmer - convert words to root words
        text=" ".join(text)
    elif lang != 'en':
        text = GoogleTranslator(target='english').translate(text)
        text = str(text).lower()   #convert to lower case
        text = re.sub('\[.*?\]', '', text) # remove punctuations
        text = re.sub('@[^\s]+','',text)  #remove tagged user
        text = re.sub('https?://\S+|www\.\S+', '', text) #remove URLs
        text = re.sub('<.*?>+', '', text) 
        text = re.sub('[%s]' % re.escape(string.punctuation), '', text)
        text = re.sub('\n', '', text)
        text = re.sub('\w*\d\w*', '', text)
        text = [word for word in text.split(' ') if word not in stopword] # word toeknization and removal of stop words
        text=" ".join(text) #join words again to form string
        text = [stemmer.stem(word) for word in text.split(' ')] # stemmer - convert words to root words
        text=" ".join(text)
        
    return text
```

```python
clean_rev_main = []
for index,row in df.iterrows():
    clean_rev = []
    for rev in range(len(row['Reviews'])):
        clean_re = clean(row['Reviews'][rev])
        clean_rev.append(clean_re)
    clean_rev_main.append(clean_rev)
df['clean_rev'] = clean_rev_main
```

```python
# extract brands from product name

from nltk.tokenize import word_tokenize

laptop_brand = ['asus','hp','apple','lenovo','dell']
shoe_brand = ['adidas','puma','nike','brooks','asics']
backpack_brand = ['nike','adidas','matein','lovevook','amazon']

df['Product_Name'] = df['Product_Name'].str.lower()
df['pname_token'] = df.apply(lambda row: nltk.word_tokenize(row['Product_Name']), axis=1)

brand_list = []
for i in range(len(df['pname_token'])):
#     print(df['pname_token'][i])
    brand = ''
    for j in range(len(df['pname_token'][i])):
        if df['pname_token'][i][j] in laptop_brand:
            brand = df['pname_token'][i][j]
        elif df['pname_token'][i][j] in shoe_brand:
            brand = df['pname_token'][i][j]
        elif df['pname_token'][i][j] in backpack_brand:
            brand = df['pname_token'][i][j]
    brand_list.append(brand)
    
df['Brand'] = brand_list
```

```python
from textblob import TextBlob

def getPolarity(text):
    return TextBlob(text).sentiment.polarity
```

```python
clean_score_main = []
for index,row in df.iterrows():
    clean_score = []
    for rev in range(len(row['clean_rev'])):
        clean_re = getPolarity(row['clean_rev'][rev])
        clean_score.append(clean_re)
    clean_score_main.append(clean_score)
df['sentiment_score'] = clean_score_main
```
